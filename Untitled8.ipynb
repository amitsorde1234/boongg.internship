# Recommending Similar jobId

Now let's create a matrix that has the user ids on one access and the jobid on another axis. Each cell will then consist of the jobid the user clicks on job. Note there will be a lot of NaN values, because most people have not click.

```
df6 = pd.pivot_table(values ='Clicks', index ='userId',columns ='jobId', data = df5) 
```

```
df6
```

Now let's grab the user click for those two jobsid:

```

job_id1 = df6[1]
job_id2 = df6[1]
job_id1.head()
```

```
clicks = pd.DataFrame(df5.groupby('jobId')['Clicks'].mean())
clicks
```

We can then use corrwith() method to get correlations between two pandas series:

```
similar_to_job_id1 = df6.corrwith(job_id1)
similar_to_job_id2 = df6.corrwith(job_id2)
```

```
similar_to_job_id1
```


Let's clean this by removing NaN values and using a DataFrame instead of a series:

```
corr_job_id1 = pd.DataFrame(similar_to_job_id1,columns=['Correlation'])
corr_job_id1.dropna(inplace=True)
corr_job_id1
```

Now if we sort the dataframe by correlation, we should get the most similar jobs id, however note that we get some results that don't really make sense. This is because there are a lot of jobsid only clicks once by users who also click jobs_id2 (it was the most clicks jobs).

```
corr_job_id1.sort_values('Correlation',ascending=False).head(10)
```

```

corr_jobid = corr_job_id1.join(clicks['Clicks'])
corr_jobid
```

Let's fix this by filtering out movies that have greater than 10 clicks

```
corr_jobid[corr_jobid['Clicks']<10].sort_values('Correlation',ascending=False).head()
```
